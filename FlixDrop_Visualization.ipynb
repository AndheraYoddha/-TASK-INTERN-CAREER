```python
import matplotlib.pyplot as plt
for column in df.columns:
    if df[column].dtype in ['float64', 'int64']:
        plt.hist(df[column].dropna())
        plt.title(f'Histogram for {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
```
