```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x = [-3,5,7]
y = [10,2,5]
```

```python
fig = plt.figure(figsize=(15,3))
plt.plot(x,y)

plt.xlim(0, 10)

plt.ylim(-3, 8)
plt.xlabel('Sales',color='blue')
plt.ylabel('product');
plt.title('Line Plot');
plt.suptitle('Sales Comparison', size=20, y = 1.05)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2,figsize=(8,6))

ax[0,0].plot(x,y,color='red',linestyle='dashed')
ax[0,0].set_title('Straight')
ax[1,0].set_title('Fly You Fools!')
ax[1,0].scatter(x,y)
ax[1,1].hist(y,color='blue')
ax[0,1].set_title('Cheese Pie')
ax[1,1].set_title('Frequency')
ax[0,1].pie(y)
plt.show()
plt.tight_layout()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mtcars = pd.read_csv("mtcars.csv")
```

```python
mtcars.columns
```

```python
mtcars.shape
```

```python
pd.crosstab(mtcars.gear,mtcars.cyl).plot(kind="bar");
#plt.show()
```

```python
mtcars["gear"].value_counts()
mtcars.gear.value_counts().plot(kind="pie")
plt.show()
```

```python
plt.figure(facecolor='green')
plt.axes()
plt.scatter(mtcars.mpg,mtcars.qsec,marker='*')## scatter plot of two variables
plt.title('Scatterplot',color='blue')
plt.xlabel('mpg',color='red',size=10)
plt.ylabel('qsec',color='purple',size=10)
plt.show()
```

```python
plt.hist(mtcars['mpg'])
plt.show()
```

```python
plt.boxplot(mtcars['mpg'])
plt.show()
```

```python
plt.hist(mtcars['mpg'],facecolor ="peru",edgecolor ="blue",bins =8)
plt.show()
```
